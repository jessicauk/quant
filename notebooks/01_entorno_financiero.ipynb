```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Todo ok üêºüî•")
```

```python
import yfinance as yf

# Descargar datos hist√≥ricos de Bitcoin en d√≥lares
btc = yf.download(
    "BTC-USD",        # Ticker de Bitcoin en USD en Yahoo Finance
    start="2025-10-01",
    end="2025-11-27",
    progress=False    # para que no salga la barra de progreso
)

btc.head(60)
```

```python
print("Filas y columnas:", btc.shape)
print("\nResumen estad√≠stico:")
print(btc.describe())
```

```python
plt.figure(figsize=(12, 4))
plt.plot(btc.index, btc["Close"])
plt.title("BTC/USD - Precio de cierre (2025)")
plt.xlabel("Fecha")
plt.ylabel("Precio de cierre (USD)")
plt.grid(True)
plt.show()
```

```python
btc.to_csv("../data/btc_2025.csv")
```
