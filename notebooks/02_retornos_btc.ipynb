```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

# Descargar datos históricos de Bitcoin en dólares
btc = yf.download(
    "BTC-USD",        # Ticker de Bitcoin en USD en Yahoo Finance
    start="2025-10-01",
    end="2025-11-27",
    progress=False    # para que no salga la barra de progreso
)

btc.head(62)
```

```python
btc["ret_simple"] = btc["Close"].pct_change()
btc[["Close", "ret_simple"]].head(62)
```

```python
btc["ret_simple"].describe()
```

```python
btc["ret_pct"] = btc["ret_simple"] * 100
btc[["ret_simple", "ret_pct"]].head(62)
```

```python
plt.figure(figsize=(10,4))
plt.hist(btc["ret_simple"].dropna(), bins=50)
plt.title("Histograma de retornos diarios de BTC (2025)")
plt.xlabel("Retorno diario")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
vol_diaria = btc["ret_simple"].std()
vol_anual = vol_diaria * np.sqrt(365)  # ~365 días de trading

print("Volatilidad diaria:", vol_diaria)
print("Volatilidad anualizada:", vol_anual)
```

```python
btc["ret_acum"] = (1 + btc["ret_simple"]).cumprod() - 1

plt.figure(figsize=(10,4))
plt.plot(btc.index, btc["ret_acum"])
plt.title("Retorno acumulado de BTC (2025)")
plt.xlabel("Fecha")
plt.ylabel("Retorno acumulado")
plt.grid(True)
plt.show()
```

### Resumen Día 2

- Calculé el **retorno diario simple** de BTC usando `pct_change()`.
- Convertí los retornos a porcentaje para interpretarlos mejor.
- Analicé la distribución de retornos con un **histograma**.
- Calculé una estimación de **volatilidad diaria y anualizada**.
- Construí la serie de **retorno acumulado** para ver cuánto habría rendido holdear BTC en Octubre-Noviembre 2025.
